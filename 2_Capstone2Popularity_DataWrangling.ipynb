```python
import pandas as pd
import numpy as np
import os
```

```python
audiofeatures_path = 'audiofeatures_dataset.csv'
spotifytracks_path = 'spotifytracks_cleaned.csv'
```

```python
df1 = pd.read_csv('audiofeatures_dataset.csv')
df2 = pd.read_csv('spotifytracks_cleaned.csv')
```

```python
merged_df = pd.merge(df1, df2, on='track_id', how='left')
```

```python
print("Merged DataFrame shape:", merged_df.shape)
merged_df.head()
```

```python
merged_df.dtypes
```

```python
merged_df['explicit'] = merged_df['explicit'].map({'True': True, 'False': False})
```

```python
columns_to_rename = {}

for col in merged_df.columns:
    if col.endswith('_x'):
        base = col.replace('_x', '')
        alt = base + '_y'

        # If both _x and _y exist, and they're equal or _x has fewer nulls, keep _x
        if alt in merged_df.columns:
            # Choose which to keep (this prefers _x)
            merged_df[base] = merged_df[col].combine_first(merged_df[alt])
            # Mark both for removal after renaming
            columns_to_rename[col] = None
            columns_to_rename[alt] = None

# Drop the old _x and _y versions
merged_df.drop(columns=[col for col in columns_to_rename if col in merged_df.columns], inplace=True)

print("✅ Cleaned up '_x' and '_y' column conflicts.")
```

```python
missing_values = merged_df.isnull().sum()
print("🔍 Missing values:\n", missing_values[missing_values > 0])
```

```python
print("Unique values per column:\n")
print(merged_df.nunique().sort_values(ascending=True))
```

```python
duplicates = merged_df.duplicated(subset='track_id')
print(f"Duplicate track_id entries: {duplicates.sum()}")
merged_df.drop_duplicates(subset='track_id', keep='first')
merged_df.dropna(subset=['danceability', 'tempo', 'energy'], inplace=True)
```

```python
merged_df.head()
```

```python
# Select only numerical columns
num_cols = merged_df.select_dtypes(include='float64').columns

# Drop rows where any of the numeric features are NaN
merged_df.dropna(subset=num_cols, inplace=True)

print(f"After dropping NaNs, final shape: {merged_df.shape}")
```

```python
merged_df.nunique().sort_values()
```

```python
merged_df.drop(columns=['explicit', 'track_index'], inplace=True)
```
