```python
corr = df.select_dtypes(include='number').corr()
top_corr_features = corr['popularity'].abs().sort_values(ascending=False)[1:6].index.tolist()

# Subset the DataFrame for pairplot
pairplot_data = df[top_corr_features + ['popularity']]
highlight_color = "#FF6F61"  

# Generate the pairplot
pairplot = sns.pairplot(pairplot_data)

# Highlight scatter plots that involve popularity
for i, feature in enumerate(top_corr_features + ['popularity']):
    for j, sub_feature in enumerate(top_corr_features + ['popularity']):
        ax = pairplot.axes[i, j]
        if 'popularity' in [feature, sub_feature] and feature != sub_feature:
            for artist in ax.collections:
                artist.set_color(highlight_color)

plt.suptitle("Pairplot: Top Correlated Features with Popularity (Highlighted)", y=1.02)
plt.tight_layout()
plt.show()
```

This pairplot visualizes the five features most strongly correlated with popularity: danceability, energy, valence, tempo, and loudness. The scatter plots involving popularity are highlighted in red to emphasize their importance.

These plots help identify patterns and trendsâ€”such as positive or negative relationshipsâ€”that can inform feature selection and modeling decisions. 

```python
#Loudness vs Popularity grouped by Loudness Level (Low, Medium, High)

# Bin loudness values
loudness_bins = [-60, -20, -10, 0]  # Decibel ranges typically seen in music
loudness_labels = ['Low', 'Medium', 'High']
df['loudness_binned'] = pd.cut(df['loudness'], bins=loudness_bins, labels=loudness_labels)

# Step 2: Plot the scatter plot with brighter colors and standard circles
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='loudness',
    y='popularity',
    hue='loudness_binned',
    palette='bright',
    alpha=0.9,
    s=70,
    marker='o'  # Use standard filled circle
)

# Step 3: Plot formatting
plt.title("Loudness vs Popularity (Grouped by Loudness Level)")
plt.xlabel("Loudness (dB)")
plt.ylabel("Popularity")
plt.grid(True)
plt.tight_layout()
plt.show()
```

This scatter plot shows how track loudness relates to popularity, using three grouped loudness levels:

- Low (Blue): â€“60 dB to â€“20 dB â†’ Very quiet tracks

- Medium (Orange): â€“20 dB to â€“10 dB â†’ Moderately loud tracks

- High (Green): â€“10 dB to 0 dB â†’ Loud tracks, near maximum volume

Each point represents a track, and color helps distinguish how different loudness levels distribute across popularity. This insight helps determine if loudness should be considered a meaningful feature in clustering or modeling.

```python
from scipy.stats import ttest_ind

key_features = ['danceability', 'loudness', 'instrumentalness', 'valence']
ttest_results = {}

# Hypothesis: High vs. Low [feature] differ in track popularity
for feature in key_features:
    if df[feature].nunique() > 1:
        median_value = df[feature].median()

        # Split into high/low groups based on feature value
        high_group = df[df[feature] > median_value]['track_popularity']
        low_group  = df[df[feature] <= median_value]['track_popularity']

        # Perform independent t-test
        stat, p = ttest_ind(high_group, low_group, nan_policy='omit')

        # Store result using scientific notation for precision
        ttest_results[feature] = format(p, ".4e")  # e.g., 2.34e-08

# Display results
print("T-test results (p-values):")
for feature, p_value in ttest_results.items():
    print(f"{feature}: p-value = {p_value}")
```

##### All 3 features have statistically significant effects on tracks popularity which are vaulable predictors. 

- Danceability
Null Hypothesis (Hâ‚€):
There is no difference in popularity between songs with high vs. low danceability.

- Alternative Hypothesis (Hâ‚):
There is a significant difference in popularity between songs with high and low danceability.

p-value: 8.3690e-21 â†’ Highly significant

Tracks with higher danceability tend to have significantly different (and likely higher) popularity than less danceable tracks.

- Loudness
Hâ‚€: Loudness level has no effect on popularity.

Hâ‚: Loudness level significantly affects popularity.

p-value: 3.8207e-09 â†’ Highly significant

Louder tracks are significantly more popular on average than quieter ones.

- Instrumentalness
Hâ‚€: Instrumentalness has no impact on popularity.

Hâ‚: Instrumentalness significantly affects popularity.

p-value: 3.6525e-13 â†’ Highly significant

Instrumental tracks tend to be less popular than vocal tracks (as shown by the negative correlation earlier), and this difference is statistically significant.
